```python
#筛选股价大于1000
import pandas as pd
import numpy as np
df = pd.DataFrame({
    'stock':['AAPL','GOOG','MSFT','AAPL','GOOG'],
    'price':[150, 2800, 300, 155, 2820],
    'volume':[1e6,2e5,5e5,1.2e6,2.2e5],
    'sector':['Tech','Tech','Tech','Tech','Tech']
})
# 筛选行
mask = df['price'] > 1000
print(mask)
# 应用筛选 + 选列
result = df.loc[mask]
print(result)
```

```python
aaa=df.query('price>1000').iloc[:,:]
print(aaa)
```
