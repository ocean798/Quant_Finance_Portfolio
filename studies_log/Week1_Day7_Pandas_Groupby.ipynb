```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'date': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02'] * 2,
    'stock': ['AAPL', 'GOOG', 'AAPL', 'GOOG', 'MSFT', 'MSFT', 'AMZN', 'AMZN'],
    'price': np.random.uniform(100, 500, 8).round(2),
    'volume': np.random.randint(1e6, 1e7, 8)
})
```

```python
# Task1: Groupby
#Groupby Stock, find price mean
print(df.groupby('stock')['price'].mean())
#Groupby date, find volume sum
print("------------")
print(df.groupby('date')['volume'].sum())
#groupby stock, find price mean, max, min, using .agg()
print("------------")
b=df.groupby('stock').agg({
    'price': ['mean','max','min'],
    'volume':'sum'
})
print(b)
```

```python
#Task 2:Merge
df1 = pd.DataFrame({
    'stock': ['AAPL', 'GOOG', 'MSFT'],
    'sector': ['Tech', 'Tech', 'Tech']
})

df2 = pd.DataFrame({
    'stock': ['AAPL', 'GOOG', 'AMZN'],
    'market_cap': [2.5, 1.8, 1.6]  # 单位：万亿
})
#inner join 内连接
print(pd.merge(df1,df2,on = 'stock', how ='inner'))
#left join 左连接
print("--------")
print(pd.merge(df1,df2,on = 'stock', how ='left'))
#right join 右连接
print("--------")
print(pd.merge(df1,df2,on = 'stock', how ='right'))
#outer join 外连接
print("--------")
print(pd.merge(df1,df2,on = 'stock', how ='outer'))
print("--------")
```

```python
#Task3:concat
q1 = pd.DataFrame({
    'stock': ['AAPL', 'GOOG'],
    'return': [0.05, 0.03]
})

q2 = pd.DataFrame({
    'stock': ['MSFT', 'AMZN'],
    'return': [0.04, 0.06]
})

q3 = pd.DataFrame({
    'stock': ['AAPL', 'GOOG'],
    'return': [0.02, 0.01]
})
print("q1")
print(q1)
print("q2")
print(q2)
print("q3")
print(q3)
print("------")
#Vertically concatenate
print(pd.concat([q1,q2],axis=0))
#Horizontal Concatenate
print(pd.concat([q1,q3],axis=1))
#Horizontal Concatenate, eliminate repeated column
print(pd.concat([q1,q3['return']],axis=1))
```

```python
#Self-assessment
import pandas as pd
sales = pd.DataFrame({
    'date': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02'],
    'product': ['A', 'B', 'A', 'B'],
    'amount': [100, 200, 150, 250],
    'store': ['X', 'X', 'Y', 'Y']
})
products = pd.DataFrame({
    'product': ['A', 'B', 'C'],
    'category': ['Electronics', 'Furniture', 'Electronics'],
    'price': [500, 300, 400]
})
q1_sales = pd.DataFrame({
    'product': ['A', 'B'],
    'q1_revenue': [5000, 3000]
})
q2_sales = pd.DataFrame({
    'product': ['A', 'B'],
    'q2_revenue': [6000, 3500]
})
#1.按 product 分组，计算每个产品的总 amount
print(sales.groupby('product')['amount'].sum())
#2.按 store 分组，计算每家店的平均 amount
print(sales.groupby('store')['amount'].mean())
#3.用 merge 把 sales 和 products 连接起来（保留所有销售记录）
print(pd.merge(sales,products,how = 'outer'))
#4.用 concat 把 q1_sales 和 q2_sales 纵向拼接
print(pd.concat([q1_sales,q2_sales],axis=0))
```
