```python
#新增
df = pd.DataFrame({
    'product': ['A', 'B', 'C', 'A', 'B'],
    'sales': [100, 200, 150, 300, 250],
    'region': ['North', 'South', 'East', 'North', 'South']
})
df['bonus'] = 0.1 * df['sales'] 
df['high_sales'] = df['sales'] >200

```

```python
df = pd.DataFrame({
    'product': ['A', 'B', 'C', 'A', 'B'],
    'sales': [100, 200, 150, 300, 250],
    'region': ['North', 'South', 'East', 'North', 'South']
})
print(df)
# 删除单列（修改原df）
a = df.drop('region', axis=1, inplace=True)
print(a)
print('------')
# 删除多列
df = pd.DataFrame({
    'product': ['A', 'B', 'C', 'A', 'B'],
    'sales': [100, 200, 150, 300, 250],
    'region': ['North', 'South', 'East', 'North', 'South']
})
df.drop(['region', 'product'], axis=1, inplace=True)
print(df)
```

```python
#排序
df = pd.DataFrame({
    'product': ['A', 'B', 'C', 'A', 'B'],
    'sales': [100, 200, 150, 300, 250],
    'region': ['North', 'South', 'East', 'North', 'South']
})
print(df.sort_values('sales', ascending = False))
print(df.sort_values(['region','sales'],ascending = [True, False]))
```
