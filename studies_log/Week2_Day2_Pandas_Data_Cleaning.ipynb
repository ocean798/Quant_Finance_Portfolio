```python
import pandas as pd
import numpy as np
#Review Questions
df = pd.DataFrame({
    'date': pd.date_range('2024-01-01', periods=10, freq='D'),
    'stock': ['AAPL', 'GOOG', 'MSFT', 'AAPL', 'GOOG', 'MSFT', 'AAPL', 'GOOG', 'MSFT', 'AAPL'],
    'price': [150, 2800, 300, 155, 2820, 310, 152, 2850, 305, 153],
    'volume': [1e6, 2e6, 5e5, 1.2e6, 2.2e6, 2.5e6, 1.1e6, 2.3e6, 3e6, 1.1e6]
})
#将date设为index
df.set_index('date', inplace = True)
#筛选stock为'AAPL'
df.query("stock == 'AAPL'")
df['turnover'] = df['price'] * df['volume']
#按stock分组，计算每组price的平均值
df.groupby('stock')['price'].mean()
#按price降序排列
df.sort_values('price',ascending = False)
#计算price日收益率
df['return'] = df['price'] / df['price'].shift() -1
#计算price的三日滚动average
df['MA3'] = df['price'].rolling(window = 3).mean()
#对周重抽样,计算周平均和gross
df.resample('W').agg({
    'price':['mean'],
    'volume':['sum']
})
# 用 apply 对 price 列应用自定义函数：计算每行的价格相对于全列均值的倍数（price / price.mean()）
df['price'].apply(lambda x: x / df['price'].mean())
#用 apply 对每行操作：新增一列 value_category，如果 turnover > 平均值则为 'High'，否则 'Low'
def compare(x):
    if x > df['turnover'].mean():
        return 'High'
    else:
        return 'Low'
df['value_category'] = df['turnover'].apply(compare)
```
