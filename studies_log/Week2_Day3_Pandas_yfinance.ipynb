```python
import pandas as pd
import numpy as np

#Practice Questions
df = pd.DataFrame({
    'date': pd.date_range('2024-01-01', periods=8, freq='D'),
    'stock': ['AAPL', 'GOOG', 'MSFT', 'AAPL', 'GOOG', 'MSFT', 'AAPL', 'GOOG'],
    'price': [150, 2800, 300, 155, 2820, np.nan, 152, 2850],
    'volume': [1e6, 2e6, 5e5, 1.2e6, np.nan, 2.5e6, 1.1e6, 2.3e6]
})
#检查每列缺失值数量
df.isna().sum()
#用前一个值填充缺失的price
df.ffill()
#删除volume缺失的行
df.dropna(how = 'any')
#将date设为索引
df.set_index('date', inplace= True)
#筛选出'stock'为'AAPL'的行
#method1
df.query("stock == 'AAPL'")
# method2
mask = df['stock'] == 'AAPL'
df[mask]
#新增列turnover=price*volume
df['turnover'] = df['price'] * df['volume']
#按stock分组，计算魅族price的均值和总合
#method1
df.groupby('stock')['price'].agg(['sum','mean'])
#method2
df.groupby('stock').agg({
    'price':['sum','mean']
})
#计算price的三日滚动均值
df['price'].rolling(window = 3).mean()
#计算日收益率
df['return'] = df['price']/df['price'].shift(1)
```

```python
import yfinance as yf
#下载AAPL 2024年数据
aapl = yf.download('AAPL',start = '2024-01-01', end = '2024-12-31')
aapl.reset_index(inplace = True)
#计算日收益率
aapl['return'] = aapl['Close'] / aapl['Close'].shift(1) - 1
print(aapl)
#计算20日滚动均值(MA20)和20日滚动波动率(年化)
aapl['MA20'] = aapl['Close'].rolling(20).mean()
aapl['vol_20'] = aapl['return'].rolling(20).std() * np.sqrt(251)
#计算每日振幅
aapl['daily_range'] = ( aapl['High'] - aapl['Low'] )/ aapl['Close']
```
