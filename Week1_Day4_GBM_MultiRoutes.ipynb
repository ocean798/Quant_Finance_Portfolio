```python
import numpy as np
import matplotlib.pyplot as plt
#初始条件
S0=100
T=252
dt=1/T
mu=0.05
sigma=0.2
N=1000
#生成所有随机数，放在矩阵dW里
dW=np.random.normal((mu-sigma**2/2)*dt,np.sqrt(dt)*sigma,(N,T))
Trend=np.exp(np.cumsum(dW,axis=1))
prices=np.zeros((N,T+1))
prices[0:,0]=S0
prices[0:,1:]=Trend[0:,0:]*S0
Mean=np.mean(prices,axis=0)
x=range(T+1)
y=Mean
plt.plot(x,y)
plt.title("Random Walk - Price Path:1000 Path Average")
plt.xlabel("Day")
plt.ylabel("Stock Price")
plt.grid(True)
plt.show()
```
