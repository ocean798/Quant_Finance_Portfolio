```python
import numpy as np
import matplotlib.pyplot as plt
#随机游走假设
S0=100
T=252
#μ是假设的市场趋势，σ^2体现出波动
dt=1/T
mu=0.05
sigma=0.2
#跌涨幅为正态分布，公式：S_t=lnS_(t-1)+exp{σϵ_t+(μ -σ^2/2)}
Trend=np.random.normal((mu-sigma**2/2)*dt,sigma*np.sqrt(dt),T)
prices=np.zeros(T+1)
prices[0]=S0
prices[1:]=S0*np.exp(np.cumsum(Trend))
#90%置信区间
U=S0*np.exp((mu-sigma**2/2)*np.arange(0,1+dt,dt)+1.645*sigma*np.sqrt(np.arange(0,1+dt,dt)))
L=S0*np.exp((mu-sigma**2/2)*np.arange(0,1+dt,dt)-1.645*sigma*np.sqrt(np.arange(0,1+dt,dt)))
#plot股价&置信区间against时间
x=range(T+1)
y=prices
plt.plot(x,y)
plt.plot(x,U)
plt.plot(x,L)
plt.title("Random Walk - Price Path")
plt.xlabel("Day")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```
